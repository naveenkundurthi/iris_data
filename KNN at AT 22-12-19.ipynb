## 4: identify the categorical in plots

```python
def plot_iris(iris, col1, col2):
    
    import seaborn as sns
    import matplotlib.pyplot as plt
    
    sns.lmplot(x= col1,y=col2,
             data=iris,
             hue="Species",
             fit_reg = False)
    
    plt.xlabel(col1)
    plt.ylabel(col2)
    
    plt.title("iris species")
    
    plt.show()
```

```python
plot_iris(iris.data, "Petal.Width", "Sepal.Length")
plot_iris(iris.data, "Sepal.Width", "Sepal.Length")
```

## data preparation

```python
from sklearn.preprocessing import scale
import pandas as pd
```

```python
colnames = ['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']
```

```python
iris_scaled = scale(iris.data[colnames])
print(iris_scaled)
```

```python
iris_scaled = pd.DataFrame(iris_scaled, columns = colnames)
```

```python
iris_scaled.head()
```

```python
iris_scaled.describe().round()
```

## 6: Numerization of the labels

```python
levels = {'setosa':0,'versicolor':1,'virginica':2}
```

```python
levels["versicolor"]
```

```python
iris_scaled['Species']=[levels[x] for x in iris.data['Species']]
iris_scaled.tail()
```

## splitting data into train and test set

```python
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
np.random.seed(123)
```

```python
m = np.asmatrix(iris_scaled)
m
```

```python
iris_split = train_test_split(m, test_size=75)
```

```python
iris_split
```

```python
x_train = iris_split[0][:,:4]
y_train = np.ravel(iris_split[0][:,4])
```

```python
x_test = iris_split[1][:,:4]
y_test = np.ravel(iris_split[1][:,4])
```

```python
print(x_train.shape)
print(x_test.shape)
```

## 8: training and evaluation of machine learning model using knn 

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
knn.fit(x_train, y_train)
```

```python
iris_test = pd.DataFrame(x_test, columns=colnames)
```

```python
iris_test.head()
```

```python
iris_test['predicted']=knn.predict(x_test)
```

```python
iris_test.head()
```

```python
iris_test['correct'] = [1 if x==z else 0 for x, z in zip(iris_test['predicted'],y_test)]
```

```python
iris_test.sample(10)
```

```python
accuracy = 100 * (sum(iris_test['correct'])/iris_test.shape[0])
```
